```python
import numpy as np
import pandas as pd
```

```python
def f_1(x1,x2,x3,x4):
  return 97*x1 + 32*x2 + 25*x3 + 20*x4 <= 139

def f_2(x1,x2,x3,x4):
  return 2*x1 + x2 + x3 + x4 <= 3

def f_3(x1,x2,x3,x4):
  r_1 = (x1 + x2 + x3 <= 2)
  r_2 = (x1 + x2 + x4 <= 2)
  r_3 = (x1 + x3 + x4 <= 2)
  return r_1 and r_2 and r_3
```

```python
import itertools
lst = list(itertools.product([0, 1], repeat=4))
res = {}
res["x"] = []
res["P1"] = []
res["P2"] = []
res["P3"] = []
for x in lst:
  res["x"].append(x)
  res["P1"].append(f_1(*x))
  res["P2"].append(f_2(*x))
  res["P3"].append(f_3(*x))
res
```

```python
df = pd.DataFrame.from_dict(res)
df
```
